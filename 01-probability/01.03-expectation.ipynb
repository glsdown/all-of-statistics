# 1.3 Expectation

## Knowledge

### Expectation of a Random Variable

The **expected value**, or **mean** or **first moment** of a random variable $X$ is defined to be:

$$
\mathbb{E}(X) = \int xdF(x) = 
\left\{\begin{array}\\
\sum_xxf(x) &= \text{if} \; X \; \text{is discrete} \\
\int xf(x)dx &= \text{if} \; X \; \text{is continuous} \\
\end{array}\right.
$$

assuming that the sum (or integral) is well defined. We use the following notation to denote the expected value of $X$:

$$
\mathbb{E}(X) = \mathbb{E}X = \int xdF(x) = \mu = \mu_X
$$

Using the **rule of the Lazy Statistician**, if $Y = r(X)$, then:

$$
\mathbb{E}(Y) = \mathbb{E}(r(X)) = \int r(x)dF_X(x)
$$

Functions of several variables are handled in a similar way. If $Z = r(X, Y)$ then:

$$
\mathbb{E}(Z) = \mathbb{E}(r(X, Y)) = \int\int r(x,y)dF(x,y)
$$

The **$k$th moment** of $X$ is defined to be $\mathbb{E}(X^k)$ assuming that $\mathbb{E}(|X|^k) < \infty$. It can be shown that if the $k$th moment exists, then for $j<k$, the $j$th moment also exists.

The **$k$th central moment** is defined to be $\mathbb{E}((X-\mu)^k)$.

### Properites of Expectations

If $X_1, \dots, X_n$ are random variables, and $a_1, \dots, a_n$ are constants, then

$$
\mathbb{E}\left(\sum_i a_i X_i\right) = \sum_i a_i \mathbb{E}(X_i)
$$

If $X_1, \dots, X_n$ are *independent* random variables, then

$$
\mathbb{E}\left(\prod_{i=1}^n X_i\right) = \prod_i \mathbb{E}(X_i)
$$

### Variance and Covariance

Let $X$ be a random variable with mean $\mu$. The **variance** (spread) of $X$ - denoted by $\sigma^2$ or $\sigma_X^2$ or $\mathbb{V}(X)$ or $\mathbb{V}X$ - is defined by

$$
\sigma^2 = \mathbb{E}[(X-\mu)^2] = \int (x-\mu)^2 dF(x)
$$

The **standard deviation** is $\text{sd}(X) = \sqrt{\mathbb{V}(X)}$ and is also denoted by $\sigma$ and $\sigma_X$.

As long as the variance is well-defined, it has the following properties:

1. $\mathbb{V}(X) = \mathbb{E}(X^2) - \mu^2$
2. If $a$ and $b$ are constants, then $\mathbb{V}(aX + b) = a^2\mathbb{V}(X)$.
3. If $X_1, \dots, X_n$ are independent, and $a_1, \dots, a_n$ are constants, then
$$
\mathbb{V}\left( \sum_{i=1}^n a_i X_i\right) = \sum_{i=1}^n a_i^2 \mathbb{V}(X_i)
$$

If $X_1, \dots, X_n$ are random variables, then we define the **sample mean** to be

$$
\bar{X_n} = \frac{1}{n}\sum_{i=1}^n X_i
$$

and the **sample variance** to be

$$
S_n^2 = \frac{1}{n-1}\sum_{i=1}^n (X_i - \bar{X_n})^2
$$

If $X_1, \dots, X_n$ are IID and $\mu = \mathbb{E}(X_i)$, $\sigma^2 = \mathbb{V}(X_i)$ then

$$
\mathbb{E}(\bar{X_n}) = \mu, \qquad \mathbb{V}(\bar{X_n}) = \frac{\sigma^2}{n} \qquad \text{and} \qquad \mathbb{E}(S_n^2) = \sigma^2
$$

Let $X$ and $Y$ be random variables with means $\mu_X$, $\mu_Y$ and standard deviations $\sigma_X$, $\sigma_Y$. Define the **covariance** (how strong the linear relationship between $X$ and $Y$ is) as

$$
\text{Cov}(X,Y) = \mathbb{E}\left( (X - \mu_X)(Y - \mu_Y)\right)
$$

and the **correlation** by

$$
\rho = \rho_{X,Y} = \rho(X, Y) = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
$$

The covariance satisfies:

$$
\text{Cov}(X, Y) = \mathbb{E}(XY) - \mathbb{E}(X)\mathbb{E}(Y)
$$

The correlation satisfies:

$$
-1 \leq \rho(X, Y) \leq 1
$$

If $Y = aX + b$ for some constants $a$, $b$, then $\rho(X, Y) = 1$ if $a > 0$ and $\rho(X, Y) = -1$ if $a < 0$. If $X$ and $Y$ are independent, then $\text{Cov}(X, Y) = \rho = 0$.

The variance and covariance have the following relationships:

$$
\mathbb{V}(X + Y) = \mathbb{V}(X) + \mathbb{V}(Y) + 2\text{Cov}(X, Y) \qquad \text{and} \qquad
\mathbb{V}(X - Y) = \mathbb{V}(X) + \mathbb{V}(Y) - 2\text{Cov}(X, Y)
$$

More generally for random variables $X_1, \dots, X_n$ we have

$$
\mathbb{V}\left( \sum_i a_i X_i\right) = \sum_i a_i^2 \mathbb{V}(X_i) + 2 \sum \sum_{i < j} a_ia_j\text{Cov}(X_i, X_j)
$$

For *multivariate* models with a random vector $X$ of the form:

$$
X = \left( 
\begin{array}\\
X_1 \\
\vdots \\
X_k \\
\end{array}
\right)
$$

The mean of the random vector $X$ is defined by

$$
\mu = \left( 
\begin{array}\\
\mu_1 \\
\vdots \\
\mu_k \\
\end{array}
\right)
=
\left( 
\begin{array}\\
\mathbb{E}(X_1) \\
\vdots \\
\mathbb{E}(X_k) \\
\end{array}
\right)
$$

The **variance-covariance matrix** $\Sigma$ is defined to be:

$$
\mathbb{V}(X) = \left[ 
\begin{array}\\
\mathbb{V}(X_1) &\text{Cov}(X_1, X_2) &\dots &\text{Cov}(X_1, X_k) \\
\text{Cov}(X_2, X_1) &\mathbb{V}(X_2) &\dots &\text{Cov}(X_2, X_k) \\
\vdots &\vdots &\vdots &\vdots \\
\text{Cov}(X_k, X_1) &\text{Cov}(X_k, X_2) &\dots &\mathbb{V}(X_k)\\
\end{array}
\right]
$$

If $a$ is a vector and $X$ is a random vector with mean $\mu$, and variance $\Sigma$, then $\mathbb{E}(a^TX) = a^T \mu$ and $\mathbb{V}(a^TX) = a^T \sum a$. If $A$ is a matrix, then $\mathbb{E}(AX) = A\mu$ and $\mathbb{V}(AX) = A \sum A^T$.

### Expectation of a Random Variable

The **conditional expectation** of $X$ given $Y = y$ is defined to be:

$$
\mathbb{E}(X|Y = y) = 
\left\{\begin{array}\\
\sum xf_{X|Y}(x|y) &= \text{discrete case} \\
\int xf_{X|Y}(x|y)dx &= \text{continuous case} \\
\end{array}\right.
$$

Since the conditional expectation is defined as a *function* of the observed value, we denote it as $\mu(y) = \mathbb{E}(X | Y = y)$.

If $r(x,y)$ is a function of $x$ and $y$ then

$$
\mathbb{E}(r(X, Y)|Y = y) = 
\left\{\begin{array}\\
\sum r(x,y)f_{X|Y}(x|y) &= \text{discrete case} \\
\int r(x,y)f_{X|Y}(x|y)dx &= \text{continuous case} \\
\end{array}\right.
$$

The **rule of Iterated Expectations** states that for random variables $X$ and $Y$, assuming the expectations exist, we have that

$$
\mathbb{E}[\mathbb{E}(Y|X)] = \mathbb{E}(Y) \qquad \text{and} \qquad \mathbb{E}[\mathbb{E}(X|Y)] = \mathbb{E}(X)
$$

More generally for any function $r(x,y)$ we have:

$$
\mathbb{E}[\mathbb{E}(r(X, Y)|X)] = \mathbb{E}(r(X,Y))
$$

The **conditional variance** is defined as

$$
\mathbb{V}(Y | X = x) = \int (y - \mu(x))^2 f(y | x)dy
$$

For random variables $X$ and $Y$, we have

$$
\mathbb{V}(Y) = \mathbb{E}\mathbb{V}(Y | X) + \mathbb{V}\mathbb{E}(Y | X)
$$

### Moment Generating Functions

The **moment generating function** MGF, or **Laplace transform** of $x$ is defined by

$$
\psi_X(t) = \mathbb{E}(e^{tX}) = \int e^{tx}dF(x)
$$

where $t$ varies over the real numbers.

If we assume the MGF is well-defined for all $t$ in some open interval around $t=0$, we have that $\psi'(0) = \mathbb{E}(X)$, and more generally taking the $k$th derivative we have that $\psi^{(k)}(0) = \mathbb{E}(X^k)$ (i.e. the $k$th moment).

The MGF has the following properties:

1. If $Y = aX + b$, then $\psi_Y(t) = e^{bt}\psi_X(at)$
2. If $X_1, \dots, X_n$ are independent and $Y = \sum_i X_i$, then $\psi_Y(t) = \prod_i \psi_{X_i}(t)$.
3. If $\psi_X(t) = \psi_Y(t)$ for all $t$ in an open interval around $0$, then $X \stackrel{d}{=} Y$ ($X$ and $Y$ are equal in distribution i.e. they have the same distribution function).

| Distribution | <div style="width:250px">MGF $\psi(t)$ </div>|
| :-- | :-: |
| Bernoulli($p$) | $pe^t + (1-p)$ |
| Binomial($n$, $p$) | $(pe^t + (1-p))^n$ |
| Poisson($\lambda$) | $\exp\left\{\lambda(e^t - 1)\right\}$ |
| Normal($\mu$, $\sigma$) | $\exp\left\{ \mu t + \frac{\sigma^2 t^2}{2}\right\}$ |
| Gamma($\alpha$, $\beta$) | $\left(\frac{1}{1-\beta t}\right)^\alpha$ for $t < \frac{1}{\beta}$ |

### Exercise 1

Let $X_1, \dots, X_n$ be $N(0,1)$ random variables and let $\bar{X_n} = n^{-1}\sum_{i-1}^n X_i$. Plot $\bar{X_n}$ vs $n$ for $n = 1, \dots, 10000$. Repeat for $X_1, \dots, X_n \sim \text{Cauchy}$. Explain why there is such a difference.

> The Cauchy one doesn't converge, whilst the Normal one does. This is because the expectation of the Cauchy distribution is not well-defined, whilst it is for Normal distribution (and is $\mu$).

```python
number_samples = 10000
n = np.arange(number_samples) + 1
```

```python
# Generate the 10000 samples
x = np.random.standard_normal(size=number_samples)
x
```

```python
sample_means = np.cumsum(x) / n
```

```python
f, ax = plt.subplots(figsize=(12, 4))

# Plot the expected value
ax.axhline(0, label="expected", color="red", ls="--")
# Plot the observed values
ax.plot(n, sample_means, label="observed")
# Add the labels
ax.legend()
ax.set_title("Expected value vs Calculated Sample Mean $N(0, 1)$")
ax.set_xlabel("n")
ax.set_ylabel("Sample Mean");
```

```python
# Generate the 10000 samples
x = np.random.standard_cauchy(size=number_samples)
x
```

```python
sample_means = np.cumsum(x) / n
```

```python
f, ax = plt.subplots(figsize=(12, 4))

# Plot the expected value
ax.axhline(0, label="expected", color="red", ls="--")
# Plot the observed values
ax.plot(n, sample_means, label="observed")
# Add the labels
ax.legend()
ax.set_title("Expected value vs Calculated Sample Mean $N(0, 1)$")
ax.set_xlabel("n")
ax.set_ylabel("Sample Mean");
```

### Exercise 2 - Stock Market

Let $Y_1, Y_2, \dots$ be independent random variables such that $\mathbb{P}(Y_i = 1) = \mathbb{P}(Y_i = -1) = \frac{1}{2}$. Let $X_n = \sum_{i=1}^n Y_i$. Think of $Y_i = 1$ as "the stock price increased by one dollar" and $Y_i = -1$ as "the stock price decreased by one dollar" and $X_n$ as "the value of the stock on day $n$". 

1. Find $\mathbb{E}(X_n)$ and $\mathbb{V}(X_n)$.

> $\mathbb{E}(X_n) = \mathbb{E}(\sum_{i=1}^n Y_i) = \sum_{i=1}^n \mathbb{E}(Y_i)$
>
> As they are identically distributed, we have $\sum_{i=1}^n \mathbb{E}(Y_i) = n \mathbb{E}(Y_1) = \sum y_1 f(y_1) = (1 \times \frac{1}{2}) + (-1 \times \frac{1}{2}) = 0$
>
> As they are independent we have
>
> $\mathbb{V}(X_n) = \mathbb{V}(\sum_{i=1}^n Y_i) = \sum_{i=1}^n \mathbb{V}(Y_i) = n \mathbb{V}(Y_1) = n \mathbb{E}(Y_1 - \mu)^2 = n \mathbb{E}[(Y_1)^2] = n \int y_1^2 dF_{Y_1}(y_1) = n \left(\left(1^2 \times \frac{1}{2}\right) + \left((-1)^2 \times \frac{1}{2}\right)\right) = n \times 1 = n$

2. Simulate $X_n$ and plot $X_n$ versus $n$ for $n=1, 2, \dots, 10000$. Repeat the whole simulation several times. Notice two things. First, it's easy to "see" patterns in the sequence even though it is random. Second you will find that the different runs look very different even though they were generated the same way. How do the calculations in part 1 explain the second observation.

> As the variance is $n$, that means that we expect the spread of the results to get larger as $n$ gets larger, so we expect the results to become increasingly different from $0$.

```python
# Number of 'days'
n = 10000
# Number of 'trials'
repeats = 25

# Generate the trial results
results = np.cumsum(np.random.choice([-1, 1], size=(repeats, n)), axis=1)
```

```python
# Create the figure
fig, ax = plt.subplots(figsize=(12, 4))

# X axis values
x = np.arange(n) + 1

# Plot each trial
for trial, result in enumerate(results):
    ax.plot(x, result, alpha=0.5)

# Add the labels
ax.set_title("Visualising the stock market")
ax.set_xlabel("Day")
ax.set_ylabel("Value");
```
